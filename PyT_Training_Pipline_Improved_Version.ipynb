```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/gscdit/Breast-Cancer-Detection/refs/heads/master/data.csv')
df.head()
```

```python
# You don't need the column id and Unnamed: 32
df.drop(columns=['id', 'Unnamed: 32'], inplace=True)
```

## Train Test Split

```python
# We divided into 80-20 split, 80 -> Train and 20 -> Test
x_train, x_test, y_train, y_test = train_test_split(df.iloc[:, 1:], df.iloc[:, 0], test_size=0.2)
```

## Scaling (Basic Preprocessing)

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Label Encoding

```python
encode = LabelEncoder()
y_train = encode.fit_transform(y_train)
y_test = encode.fit_transform(y_test)
```

## NumPy arrays to PyTorch Tensors

```python
x_train_tesnsor = torch.from_numpy(x_train)
x_test_tesnsor = torch.from_numpy(x_test)
y_train_tesnsor = torch.from_numpy(y_train)
y_test_tesnsor = torch.from_numpy(y_test)
```

## DataSet and DataLoader

```python
#  The Improvement using the dataset and dataloader
from torch.utils.data import Dataset, DataLoader

class custDataset(Dataset):

  def __init__(self, features, labels):
    self.feature = features
    self.label = labels

  def __len__(self):

    return len(self.feature)

  def __getitem__(self, idx):
    return self.feature[idx], self.label[idx]
```

```python
train_dataset = custDataset(x_train_tesnsor, y_train_tesnsor)
test_dataset = custDataset(x_test_tesnsor, y_test_tesnsor)
```

```python
train_dataset[10]
```

```python
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=32, shuffle=True)
```

## Define the Model (Creation of Model)

```python
class MySimpleNN(nn.Module):

  def __init__(self, num_features):

    super().__init__()
    self.linear = nn.Linear(num_features, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, x):
    out = self.linear(x)
    out = self.sigmoid(out)

    return out
```

### Important Parameters

```python
learning_rate = 0.1
epochs = 25
```

```python
# Loss Function using the nn module
loss_function = nn.BCELoss()
# Simple that is only change you have to do
```

### Training Pipline

```python
import torch.optim as optim

# create the model
model = MySimpleNN(x_train_tesnsor.shape[1])

# define optimizer
optimizer = optim.SGD(model.parameters(), lr=learning_rate)

for epoch in range(epochs):

  for batch_feature, batch_label in train_loader:

    # Forward Pass
    y_pred = model(batch_feature.float())

    # Loss Calculation
    loss = loss_function(y_pred, batch_label.view(-1, 1).float())
    # What happen first was y_pred having the shape [455] but the y_train_tensor having shap of [455, 1] so you just reshape or use the view

    # Clear Gradients
    optimizer.zero_grad()

    # Backward Pass
    # You have to calculate the derviate
    loss.backward()

    # Parameters Updation
    optimizer.step()

    # Print Loss inn epoch
    print(f"Epoch: {epoch+1}, Loss: {loss.item()}")
```
