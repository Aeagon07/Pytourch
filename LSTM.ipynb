```python
# Tokenization
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
# Tokenize
tokens = word_tokenize(document.lower())
```

```python
# Build our Vocab -> It is group of unique tokens
vocab = {'<unk>': 0}

ele = Counter(tokens) # it will create the dictionary of how many times the word is repeating
# But you have to find the unique token from it so use the keys() then you have to build the vocab so you just add those token who are not present in vocab
for token in ele.keys():
  if token not in vocab:
    vocab[token] = len(vocab)
```

```python
# extract sentence from data
input_sentence = document.split('\n')

input_sentence[0]
```

```python
# text to index
def text_to_index(sentence, vocab):

  numerical_sentence = []

  for token in sentence:
    if token in vocab:
      numerical_sentence.append(vocab[token])
    else:
      numerical_sentence.append(vocab['<unk>'])

  return numerical_sentence
```

```python
# Tokenize the sentences to the number

input_num_sent = []
for sentence in input_sentence:
  input_num_sent.append(text_to_index(word_tokenize(sentence.lower()), vocab))
  # here every sentence is tokenize means split in to the list
  # And, now we convert them to numerical and add them to the list input_num_sent
```

```python
print(input_num_sent[0]) # now we have the numerical form of the each statement in the document

print(len(input_num_sent))
```

```python
training_sequence = [] # which stores the next numerical array
for sentence in input_num_sent:

  for i in range(1, len(sentence)):
    training_sequence.append(sentence[:i+1])
```

```python
len(training_sequence)
```

```python
# find out the maximum sequence in the list
len_list = []

for sqeunce in training_sequence:
  len_list.append(len(sqeunce))

max(len_list)
```

```python
training_sequence[0] # here we have to add 60 zeros before the 1, 2
```

```python
padded_training_seq = []
for seq in training_sequence:
  padded_training_seq.append([0]*(max(len_list) - len(seq)) + seq)
```

```python
len(padded_training_seq[10])
```

```python
padded_training_seq = torch.tensor(padded_training_seq, dtype=torch.long)
```

```python
padded_training_seq.shape
```

```python
X = padded_training_seq[:, :-1]
y = padded_training_seq[:, -1]
```

```python
class CustomDataset(Dataset):
  def __init__(self, X, y):
    self.X = X
    self.y = y

  def __len__(self):
    return len(self.X)

  def __getitem__(self, index):
    return self.X[index], self.y[index]
```

```python
dataset = CustomDataset(X, y)
```

```python
len(dataset)
```

```python
dataloader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
class LSTMModel(nn.Module):

  def __init__(self, vocab_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, 100)
    self.lstm = nn.LSTM(100, 150, batch_first=True) # LSTM gives you three output so there no point of using the sequential container
    self.fc = nn.Linear(150, vocab_size)

  def forward(self, x):
    embedded = self.embedding(x)
    intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.lstm(embedded)
    output = self.fc(final_hidden_state.squeeze(0))
    return output
```

```python
model = LSTMModel(len(vocab))
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
model.to(device)
```

```python
epochs = 50
learning_rate = 0.001

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
# trainig loop
for epoch in range(epochs):
  total_loss = 0
  for batch_x, batch_y in dataloader:
    batch_x = batch_x.to(device)
    batch_y = batch_y.to(device)

    optimizer.zero_grad()

    output = model(batch_x)

    loss = criterion(output, batch_y)

    loss.backward()

    optimizer.step()

    total_loss += loss.item()

  print(f"Epoch: {epoch+1}, Loss: {total_loss/len(dataloader)}")
```

```python
# Prediction

def prediction(model, vocab, text):

  # tokenize
  tokeized_text = word_tokenize(text.lower()) # if you gievn the some line in the documnet it will show you th next words to us

  # text -> numerical indices
  numerical_text = text_to_index(tokeized_text, vocab) # after text to number it will gives me the numerical data of that predicted number

  # padding
  padded_text = torch.tensor([0] * (61 - len(numerical_text)) + numerical_text, dtype=torch.long).unsqueeze(0)
  # so we convert the each numerical ensor to the length of 61 D by adding zero infront position
  # But after that it will show you just 61 we want it show ike the batch size, number of words so, unsqueeze it

  # send to model
  output = model(padded_text)

  # predicted index
  value, index = torch.max(output, dim=1) # now here the index can gives you the next word so,

  # merge with text
  return text + " " + list(vocab.keys())[index]
```
